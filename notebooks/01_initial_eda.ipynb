# Stroke Prediction - Initial EDA
**Dataset**: Stroke Prediction Dataset (Kaggle, fedesoriano 2021)
**Tujuan**: Memahami distribusi data, missing values, dan class imbalance


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/raw/healthcare-dataset-stroke-data.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Cek distribusi target (class imbalance)
print(df['stroke'].value_counts())
print(f'Prevalensi stroke: {df['stroke'].mean()*100:.1f}%')
```

```python
# Cek missing values
print(df.isnull().sum())
# BMI ada 201 missing values - perlu imputasi dengan median
```
